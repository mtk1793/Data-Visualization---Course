```python
# Using for loops in the axes
fig, axes = plt.subplots(2, 2, figsize=(8, 8), sharex=True, sharey=True)

x = np.linspace(0, 20, 100)
y = np.sin(x)
axes[0, 1].plot(x, y)
# axes[0, 1].set_title("Upper LEft")
# axes[0, 0].set_title("Upper LEft")
# axes[1, 1].set_title("Upper LEft")
# axes[1, 0].set_title("Upper LEft")

# Using flat will make all of the 2D axes together in a line and we can do the iteration:
for ax in axes.flat:
    ax.set(xlabel="x", ylabel="f(x)")
    ax.grid()

# Add some spacing among the figures
fig.tight_layout(pad=0.5)
```

```python
# Doing grid specs
import matplotlib.gridspec as gridspec

fig = plt.figure(figsize=(8, 8), dpi=100)
gs = gridspec.GridSpec(2, 2)

ax1 = fig.add_subplot(gs[0, :])  # Top row, span all columns
ax2 = fig.add_subplot(gs[1, 0])  # Bottome left
ax3 = fig.add_subplot(gs[1, 1])  # Bottom right

x = np.linspace(0, 20, 100)

ax1.plot(x, np.sin(x))
ax2.plot(x, np.cos(x))
ax3.plot(x, np.tan(x))

# Set the facecolor of the figure

fig.patch.set_facecolor("#E6E6E6")

fig.suptitle("Grids", fontsize=18)

# Change the setting for the linestyle in grids

ax3.grid(linestyle="--", alpha=0.7, linewidth=1)

# Set x ticks based on costum number
ax3.set_xticks([0, 10, 20])
ax3.set_xticklabels(["Zero", "One", "Two"], rotation=45)
```

```python
# Using Runtime Configuration Parameters or rCParams where you can apply one setting for the whole figures

x = np.linspace(0, 20, 100)
y = np.sin(x)

# Using Style Sheets

plt.style.use("seaborn-v0_8-darkgrid")

# Use code plt.style.available to find the available styles in the figure to use

plt.rcParams["font.size"] = 12

fig, axes = plt.subplots()

axes = plt.plot(x, y)
```
