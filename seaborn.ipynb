```python
# Load the dataset

import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# Generate random data
np.random.seed(42)
data = {
    "Category": np.random.choice(["A", "B", "C", "D", "E"], size=100),
    "Values": np.random.randn(100) * 10 + 50,
}

# Convert to DataFrame
df = pd.DataFrame(data)

# Visualize the dataset using seaborn
plt.figure(figsize=(8, 6))
sns.boxplot(x="Category", y="Values", data=df)
plt.title("Random Data Box Plot")
plt.xlabel("Category")
plt.ylabel("Values")
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Best practice for loading a figure and axes for plotting the data
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 6))

# Load the dataset from the sns itself
penguins = sns.load_dataset("penguins")
penguins

# Hue is for the color encoding - use for the seperation
sns.histplot(data=penguins, x="bill_depth_mm", hue="species", ax=ax1)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

tips = sns.load_dataset("tips")
tips

sns.set_theme(font_scale=1.07)

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 5))

# scatter
sns.scatterplot(data=tips, ax=ax1, x="total_bill", y="tip", hue="time", size="size")

# boxplot
sns.boxplot(data=tips, x="day", y="total_bill", ax=ax2, hue="time")

plt.tight_layout()
```
