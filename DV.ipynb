# Simple Example

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0, 10, 100)
y = np.sin(x)

plt.plot(x, y)
plt.title("A simple sinewave")
# plt.show()
# Plotting order : Figure --> Axes --> Axis --> Artists
```

# In details 

```python
import matplotlib.pyplot as plt
import numpy as np

# Creating figure and axes
fig, ax = plt.subplots()

# fig, ax = plt.subplots(2, 2, figsize=(3, 4))

# Generating data
x = np.linspace(0, 20, 100)
y = np.sin(x)

# Plot the data on the axes
ax.plot(x, y)

# Changing the x-axis and y-axis title and lables

ax.set_xlabel("Time (s)")
ax.set_ylabel("Amplitude")
ax.set_title("A simple sinewave")
```

```python
x = np.linspace(0, 2*np.pi, 100)
y1 = np.sin(x)
y2 = np.cos(x)

fig, ax = plt.subplots(figsize=(8,4))
ax.plot(x, y1, label='sin(x)')
ax.plot(x, y2, label='cos(x)')
ax.set_xlabel('Time (s)')
ax.set_ylabel('Output')

ax.legend()
```

```python
x = np.linspace(0, 2 * np.pi, 100)
y1 = np.sin(x)
y2 = np.cos(x)

fig, ax = plt.subplots(figsize=(8, 4))
ax.scatter(x, y1, label="sin(x)")
ax.scatter(x, y2, label="cos(x)")
ax.set_xlabel("Time (s)")
ax.set_ylabel("Output")

ax.legend()
```
